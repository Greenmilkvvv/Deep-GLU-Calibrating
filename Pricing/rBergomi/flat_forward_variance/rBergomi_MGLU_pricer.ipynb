```python
import sys
sys.path.append('..')
from rBergomi_utils import *
```

## 数据读取

```python
xx, yy, strikes, maturities = data_read(r"../../../Data/rBergomiTrainSet.txt.gz")
```

```python
# 训练集的 Upper and Lower Bounds
upper_bound = np.array([0.16,4,-0.1,0.5])
lower_bound = np.array([0.01,0.3,-0.95,0.025])
```

## 训练 & 测试集 构建

```python
# 确定训练数据
data_loader, train_data, test_data = get_dataset_for_train(xx, yy, upper_bound, lower_bound, device='cuda')
```

## 训练

```python
import sys
sys.path.append(r"../../../") 

from Networks.nn import NN_pricing_MGLU
from Networks.train_utils import train_model
from torch.nn import MSELoss

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')


hyperparams = { 
    "input_dim": 4,
    'hidden_dim': 64, 
    'hidden_nums': 4, 
    'output_dim': 88
}

model = NN_pricing_MGLU(hyperparams).to(device=device, dtype=torch.float64)

loss_MSE = MSELoss()

# 优化器的特殊设置
# 分离掩码参数和其他参数
mask_params = []
other_params = []
for name, param in model.named_parameters():
    if 'mask' in name:
        mask_params.append(param)
        print(f"Mask param: {name}, shape: {param.shape}")
    else:
        other_params.append(param)

# 为不同参数组配置不同的学习率
optim_Adam = torch.optim.Adam([
    {'params': other_params, 'lr': 0.001},   # 主参数用较高学习率
    {'params': mask_params, 'lr': 0.00001}   # 掩码参数用极低学习率，建议是其他的1/100
], weight_decay=1e-5)  # 可以加入轻微权重衰减


tain_loss_lst, test_loss_lst = train_model( 
    loss_MSE, 
    optim_Adam, 
    model, 
    data_loader, 
    train_data, 
    test_data, 
    epochs=10
)

print(f"训练集损失: {tain_loss_lst[-1]}")
print(f"测试集损失: {test_loss_lst[-1]}")
```

```python
torch.save(model.state_dict(), r"../../../Results/models/MGLU_rBergomi_flat_forward_variance.pth")

print("模型已保存")
```
